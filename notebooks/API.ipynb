### Hits 2022

```python
#https://open.spotify.com/playlist/56r5qRUv3jSxADdmBkhcz7?si=y_zPxOysQn64CoXgGj3XOQ
```

```python
#Obtenemos las canciones teniendo en cuenta la paginación (incialmente solo sacaba hasta el registro 70)

playlist_id = "56r5qRUv3jSxADdmBkhcz7"
offset = 0
track_data2022 = []

while True:
    playlist_tracks = sp.playlist_tracks(playlist_id, offset=offset)
    
    if not playlist_tracks['items']:
        break
    
    for item in playlist_tracks['items']:
        track = item['track']
        
        track_info2022 = {
            'id': track['id'],
            'name': track['name'],
            'artists': ', '.join([artist['name'] for artist in track['artists']]),
            'duration_ms': track['duration_ms'],
            'release_date': track['album']['release_date'],
            'year': track['album']['release_date'][:4],
            'popularity': track['popularity'],
            'explicit': track['explicit']}
        
        track_data2022.append(track_info2022)
    
    offset += len(playlist_tracks['items'])
```

```python
track2022_ids = [track['id'] for track in track_data2022]
audio2022_features = sp.audio_features(track2022_ids)
```

```python
#Añadimos las canciones
for i, features in enumerate(audio2022_features):
    if features:
        track_data2022[i].update({
            'acousticness': features['acousticness'],
            'danceability': features['danceability'],
            'energy': features['energy'],
            'instrumentalness': features['instrumentalness'],
            'liveness': features['liveness'],
            'loudness': features['loudness'],
            'speechiness': features['speechiness'],
            'tempo': features['tempo'],
            'valence': features['valence'],
            'mode': features['mode'],
            'key': features['key']})
```

```python
df_tracks2022 = pd.DataFrame(track_data2022)

df_tracks2022
```
